## 4. Preliminary Analysis

```python
df = df[["G3","Walc","sex","romantic","studytime","absences","freetime"]]
df
```

```python
import seaborn as sns
import matplotlib.pyplot as plt

sns.pairplot(df.select_dtypes(include=['number']))
plt.suptitle("Pairs Plot of Numerical Variables")
plt.show()
```

```python
import statsmodels.formula.api as smf
full_model = smf.ols(
    formula='G3 ~ Walc + absences + freetime + studytime + sex',
    data=df).fit()

full_model.summary()
```

```python
import matplotlib.pyplot as plt

plt.scatter(full_model.fittedvalues, full_model.resid)
plt.axhline(y=0, color='r', linestyle='--')
plt.xlabel('Fitted (Predicted) values')
plt.ylabel('Residuals')
plt.title('Fitted values vs. Residuals Plot')
plt.show()
```

 I believe transforming one or more variables could improve my model’s fit. The residual plot shows a clear pattern and unequal spread, which suggests the relationship between some predictors and the final grade (G3) may not be perfectly linear. In particular, absences is likely skewed, and transforming it could help the model better meet the assumptions of linear regression and improve prediction accuracy.

```python
import numpy as np
import statsmodels.formula.api as smf

df['log_absences'] = np.log1p(df['absences'])


model_transformed = smf.ols('G3 ~ Walc + log_absences + freetime + studytime + sex',data=df).fit()
model_transformed.summary()
```

```python

plt.scatter(model_transformed.fittedvalues, model_transformed.resid)
plt.axhline(0, color='red', linestyle='--')
plt.xlabel("Fitted Values (Transformed Model)")
plt.ylabel("Residuals")
plt.title("Residuals vs. Fitted Values (After Transformation)")
plt.show()
```

After transforming absences using a log transformation(to get rid of the outlierts within absences) and refitting the model, the residual plot did not improve significantly. The residuals still show an uneven spread of residuals and a possible group of outliers around -10. This suggests that the transformation didn’t meaningfully improve the model’s fit or bring the linearity assumption closer to being met.
