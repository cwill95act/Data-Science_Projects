```python

def Mr_clean(data):
    data.columns = data.columns.str.strip()

    # Deleting NA values
    Not_clean = data.dropna(subset=['health condition.1', 
                                    'evidence score 0 = harmful \n1 = no / insufficient evidence\n6 = strong', 
                                    'popular interest'])

    # Renaming wanted columns for further simplicity
    Not_clean = Not_clean.rename(columns={
        'health condition.1': 'Category',
        'evidence score 0 = harmful \n1 = no / insufficient evidence\n6 = strong': 'Evidence Score',
        'popular interest': 'Popularity'
    })

    # Keeping all numerical values 
    Not_clean = pd.concat([
        Not_clean.select_dtypes(include=['number']),
        pd.get_dummies(Not_clean['Category'], prefix='Category', drop_first=True)
    ], axis=1)

    return Not_clean

#correlation matrix
cleaned_data = Mr_clean(data)
correlation_matrix = cleaned_data.corr(numeric_only=True)

correlation_matrix

```

cleaned data focuses on the catgorized health conditions and the corresposning evidence scores for each row repsonding to a health condtion. 

```python
#heat map

import matplotlib.pyplot as plt
import seaborn as sns

plt.figure(figsize=(24, 20))
sns.heatmap(correlation_matrix, annot=True, fmt=".2f", cmap="coolwarm", cbar=True)
plt.title("Coefficient Correlation Matrix")
plt.show()
```

```python

chosen_category = "Category_general health"

#Calculating slope and intercept
std_x = cleaned_data["Evidence Score"].std()
std_y = cleaned_data[chosen_category].std()
correlation = cleaned_data.corr(numeric_only=True).loc["Evidence Score", chosen_category]
slope = correlation * (std_y / std_x)
avg_x = cleaned_data["Evidence Score"].mean()
avg_y = cleaned_data[chosen_category].mean()
intercept = avg_y - slope * avg_x

print(f"Slope for {chosen_category}: {slope}")
print(f"intercept for {chosen_category}: {intercept}")


```

```python

import numpy as np
import matplotlib.pyplot as plt

slope_general_health = 0.06792058516196449
intercept_general_health = 0.0010449320794148204


x_general_health = np.array([4,1,2,5,1,1,2])
y_general_health = slope_general_health * x_general_health + intercept_general_health


plt.plot(x_general_health, y_general_health, label="Regression Line (General Health)", color="blue")
plt.xlabel("Evidence Score")
plt.ylabel("Correlation(predicted benefit)")
plt.title("Linear Regression: General Health")
plt.legend()
plt.grid(True)
plt.show()
```

This graph illustrates the potential cannabis has to improve general health conditions, as shown by the string positive linear relationship between the evidence score and predicted benefit. The regression line demonstrates that as the amount of evidence score supporting cannabis health benefits increases. The predicted correlation with general health is about .10 to .35. This data strengthens the argument by showing cannabis's measurable and scientifically supported impact on conditions that fall into the category of general health.
